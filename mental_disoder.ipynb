```python
df=pd.read_csv('mental_disorders_dataset.csv')
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,6)
```

```python
print(df.isnull().sum())
```

```python
df.info()
```

```python
for col in ['Sexual Activity','Concentration','Optimisim']:
    if df[col].dtype == 'object':
        df[col] = df[col].str.extract(r'(\d+)').astype(float)
```

```python
df.describe()
```

```python
sns.countplot(x='Expert Diagnose',data=df,palette='coolwarm')
plt.title('Distribution of expert diadnoses')
plt.show()
```

```python
binary_cols = ['Mood Swing','Suicidal thoughts','Anorxia','Authority Respect',
               'Try-Explanation','Aggressive Response','Ignore & Move-On',
               'Nervous Break-down','Admit Mistakes','Overthinking']
```

```python
df[col] = df[col].map({'YES': 1, 'NO': 0})
```

```python
num_cols=['sexual activity','concentration','optimism']
```

```python
sns.boxplot(data=df,x='Expert Diagnose',y=col,palette='viridis')
plt.title(f"{col} by diagnosis")
plt.show()
```

```python
corr=df.select_dtypes(include=['number']).corr()
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('correlation heatmap (numeric features)')
plt.show()
```

```python
binary_sum = df[binary_cols].sum().sort_values(ascending=True)
plt.figure(figsize=(10,6))
sns.barplot(x=binary_sum.values, y=binary_sum.index)
plt.title("Frequency of YES Responses Across Patients")
plt.xlabel("Count of YES")
plt.show()
```

```python
 plt.figure(figsize=(8,5))
sns.histplot(df[col], kde=True, bins=10, color='skyblue')
plt.title(f"Distribution of {col}")
plt.xlabel(col)
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df[df['Expert Diagnose'].isin(['Bipolar Type-1','Bipolar Type-2'])],
              x='Mood Swing', hue='Expert Diagnose', palette='coolwarm')
plt.title('Mood Swing Comparison: Type-1 vs Type-2')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='Mood Swing',hue='Expert Diagnose', data=df)
plt.title('Mood swings expert diagnose data')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Overthinking',hue='Expert Diagnose', data=df)
plt.title('Overthinking Frequency by Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x='Authority Respect',hue='Expert Diagnose', data=df)
plt.title('Authority Respect Frequency by Diagnosis')
plt.show()
```
